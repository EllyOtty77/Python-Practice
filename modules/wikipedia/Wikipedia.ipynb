```python
import wikipedia as wiki
```

```python
# Search query
results = wiki.search('Nietzsche')
results
```

```python
# Instantiate article object
niet = wiki.page(results[0])
```

```python
# Print summary 
niet.summary
```

```python
# Show title
niet.title
```

```python
# Full content
niet.content
```

```python
# Access embeded links
niet.links
```

```python
# References 
niet.references
```

```python
# Image links
niet.images
```

```python
# Specific section
niet.section('History')
```

```python
# Languages
wiki.languages()
```
